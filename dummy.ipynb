```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.tail()
```

```python
import pandas as pd
```

```python
df = pd.get_dummies(df, columns=["sex"], drop_first=True).tail()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df["pclass"] = df["pclass"].astype("category")
```

```python
df.dtypes
```

```python
fitted = OneHotEncoder(categories="auto", handle_unknown="error", sparse=False).fit(df.loc[:, ["pclass"]])
## sparse=False : array 반환
```

```python
df = pd.concat([df, pd.DataFrame(fitted.transform(df.loc[:, ["pclass"]]), columns=["c1", "c2", "c3"])], axis=1)
```
