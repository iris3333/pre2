# Descriptive Statistics for Qualitative Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Qualitative Data

```python
sandwich = pd.read_csv("Sandwich.csv")
print(sandwich.head())
```

```python
print("Bread:", set(sandwich["Bread"]))
print("Filling:", set(sandwich["Filling"]))
```

### 2. One-way Frequency Table

```python
one_freq = pd.crosstab(index=sandwich["Bread"], columns="Counts")
print(one_freq)
```

```python
print(round(one_freq/one_freq.sum(), 2))
```

```python
one_freq.plot(kind="bar",
              color=("#D7CCC8", "#BCAAA4", "#A1887F"), ylim=(0, 60), rot=0, legend=False)
plt.xlabel(""); plt.show()
```

```python
one_freq.plot(kind="pie", subplots=True, figsize=(4, 4),
              colors=("#D7CCC8", "#BCAAA4", "#A1887F"), legend=False, autopct="%.2f")
plt.ylabel(""); plt.show()
```

### 3. Two-way Frequency Table

```python
two_freq = pd.crosstab(index=sandwich["Bread"], columns=sandwich["Filling"], margins=True)
print(two_freq)
```

```python
print(round(two_freq/two_freq.loc["All", "All"], 2))
```

```python
two_freq.iloc[:-1, :-1].T.plot(kind="bar", stacked=True,
                               color=("#D7CCC8", "#BCAAA4", "#A1887F"), ylim=(0, 60), rot=0)
plt.xlabel(""); plt.show()
```

```python
two_freq.iloc[:-1, :-1].plot(kind="pie", subplots=True, figsize=(13.5, 4),
                             colors=("#D7CCC8", "#BCAAA4", "#A1887F"),
                             legend=False, labels=("", "", ""), autopct="%.2f")
plt.legend(two_freq.iloc[:-1, :-1].index, title="Bread", bbox_to_anchor=(1, 0.7)); plt.show()
```
